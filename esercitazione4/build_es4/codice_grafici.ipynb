```
import pandas as pd
import matplotlib.pyplot as plt
```

```
dt_tempi = pd.read_csv("tempi_ordinamento.csv")
dt_tempi
```

```
plt.figure(figsize=(10,6))
plt.plot(dt_tempi['Dimensione'], dt_tempi['BubbleSort'], label='Bubble Sort', marker = 'o')
plt.plot(dt_tempi['Dimensione'], dt_tempi['InsertionSort'], label='InsertionSort', marker='s')
plt.plot(dt_tempi['Dimensione'], dt_tempi['SelectionSort'], label='SelectionSort', marker='^')
plt.plot(dt_tempi['Dimensione'], dt_tempi['StdSort'], label='std::sort (IntroSort)', marker='x', linewidth=2)
plt.title('Confronto tempi di esecuzione Algoritmi di Ordinamento')
plt.xlabel('Dimensione del Vettore ($N$)')
plt.ylabel('Tempo di esecuzione (secondi)')
plt.legend()
plt.grid(True)
plt.show()
```
